```python
import pandas as pd

# Load the input dataset
df = pd.read_csv("/content/original_expensive_dataset.csv")

# Convert to required format
converted_df = pd.DataFrame({
    "Drug Code": df["id"],
    "Name": df["name"],
    "Unit Size": df["pack_size_label"],
    "Price": df["price(₹)"],
    "Group Name": df["type"],
    "Salts": df.apply(
        lambda row: row["short_composition1"] if pd.isna(row["short_composition2"])
        else f"{row['short_composition1']} + {row['short_composition2']}", axis=1
    )
})

# Save the result as expensive_preprocess_columnNameChange.csv
converted_df.to_csv("expensive_preprocess_columnNameChange.csv", index=False)
```
