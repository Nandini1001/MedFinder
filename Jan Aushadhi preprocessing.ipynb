```python
import pandas as pd

# Load the input dataset
df = pd.read_csv("/content/original_janAushadhi.csv")

# Rename columns to match reference
df = df.rename(columns={
    'Sr No': 'Sr No',
    'Drug Code': 'Drug Code',
    'Generic Name': 'Name',
    'Unit Size': 'Unit Size',
    'MRP': 'Price',
    'Group Name': 'Group Name'
})

# Drop rows where Price is missing or zero
df = df[df['Price'].notnull()]
df = df[df['Price'] != 0]

# Drop duplicate rows
df = df.drop_duplicates(subset=['Drug Code', 'Name', 'Unit Size', 'Price', 'Group Name'])

# Add empty 'Salts' column
df['Salts'] = ""

# Drop 'Sr No' column
df.drop('Sr No', axis=1, inplace=True)

# Reorder columns (without 'Sr No')
df = df[['Drug Code', 'Name', 'Unit Size', 'Price', 'Group Name', 'Salts']]

# Save and download the cleaned file
df.to_csv("janAushadhi_preprocessed.csv", index=False)
```
